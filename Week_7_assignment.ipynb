```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/StudentsPerformance.csv")
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print("\nData Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
edu_group = df.groupby("parental level of education")["math score"].mean()
print("\nAverage Math Score by Parental Education:")
print(edu_group)
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=df.index, y=df["math score"].cumsum(), label="Cumulative Math Score")
plt.title("Cumulative Math Score Over Records")
plt.xlabel("Record Index")
plt.ylabel("Cumulative Math Score")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="parental level of education", y="math score", data=df, estimator="mean", ci=None)
plt.title("Average Math Score by Parental Education")
plt.xlabel("Parental Level of Education")
plt.ylabel("Average Math Score")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["reading score"], bins=15, kde=True, color="skyblue")
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="math score", y="writing score", data=df, hue="gender", alpha=0.7)
plt.title("Math Score vs Writing Score")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")
plt.legend(title="Gender")
plt.show()
```
