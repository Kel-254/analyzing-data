```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/StudentsPerformance.csv")
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print("\nData Info:")
print(df.info())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
edu_group = df.groupby("parental level of education")["math score"].mean()
print("\nAverage Math Score by Parental Education:")
print(edu_group)
```

```python
df_sorted = df.sort_index()
plt.figure(figsize=(8,5))
plt.plot(df_sorted.index, df_sorted["math score"].cumsum(), label="Cumulative Math Score")
plt.title("Cumulative Math Score Over Records")
plt.xlabel("Record Index")
plt.ylabel("Cumulative Math Score")
plt.legend()
plt.show()
```

```python
edu_group.plot(kind="bar", figsize=(8,5))
plt.title("Average Math Score by Parental Education")
plt.xlabel("Parental Level of Education")
plt.ylabel("Average Math Score")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df["reading score"], bins=15, edgecolor="black")
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["math score"], df["writing score"], c="blue", alpha=0.6, label="Data Points")
plt.title("Math Score vs Writing Score")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")
plt.legend()
plt.show()
```
